```python
cleandata = Path("dataframev2.csv")
df_clean_data = pd.read_csv(cleandata)
df_clean_data.head()
```

```python
df_clean_data_encoded = pd.get_dummies(df_clean_data, columns = ['hom league','vis league'])
```

```python
df_clean_data_encoded.drop(columns = 'spread', inplace = True)
```

```python
df_clean_data_encoded['Date'] = df_clean_data_encoded['Date'].apply(lambda d:pd.to_datetime(d,format='%Y%m%d'))
```

```python
X = df_clean_data_encoded.copy()
X.drop(["home_winner", 'Date'], axis=1, inplace=True)
X.head()
```

```python
y = df_clean_data_encoded["home_winner"].ravel()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_scaler = scaler.fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train.isnull().sum()
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_model
```

```python
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data
predictions = rf_model.predict(X_test_scaled)
```

```python
# Calculating the confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
